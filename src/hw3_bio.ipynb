```python
import pandas as pd

data = [
        pd.read_csv('counts/SRR3414629.counts', sep="\t",  names=['geneID', 'SRR3414629']),
        pd.read_csv('counts/SRR3414630.counts', sep="\t",  names=['geneID', 'SRR3414630']),
        pd.read_csv('counts/SRR3414631.counts', sep="\t",  names=['geneID', 'SRR3414631']),
        pd.read_csv('counts/SRR3414635.counts', sep="\t",  names=['geneID', 'SRR3414635']),
        pd.read_csv('counts/SRR3414636.counts', sep="\t",  names=['geneID', 'SRR3414636']),
        pd.read_csv('counts/SRR3414637.counts', sep="\t",  names=['geneID', 'SRR3414637'])
]
result = data[0].copy()
for i in range(1, 6):
  result = result.merge(data[i], on='geneID', how='left')
result = result.set_index('geneID')
result.head()
```

```python
result.to_csv('counts/ALL.counts', sep='\t', index=True)
```
